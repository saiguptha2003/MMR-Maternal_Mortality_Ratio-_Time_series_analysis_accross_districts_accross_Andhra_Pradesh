```python
import matplotlib.pyplot as plt
import pandas as pd
# file = pd.read_excel('data.xlsx')
# x_axis = file['X values']
# y_axis = file['Y values']
# plt.bar(x_axis, y_axis, width=5)
# plt.xlabel("X-Axis")
# plt.ylabel("Y-Axis")
# plt.show()
l=['Anantapur','Chittoor','East Godavari','Guntur','Krishna','Kurnool','Prakasam','Srikakulam','Nellore','Visakhapatnam','Vizianagaram','West Godavari','Cuddapah']
color=['red','green','yellow','brown','cyan','#249571','black','#630000','#C4B4B1','#E400FF','#E4E859','#FF9E00','#A46C7C']
j=0
plt.title("graph")
for i in l:
    print()
    filename=pd.read_excel(r"/content/{}_HB.xlsx".format(i))
    x_axis=filename['year']
    y_axis=filename['HB < 7mg']
    plt.rcParams['figure.figsize']=[10,10]
    plt.plot(x_axis,y_axis,color=color[j],label=l[j])
    j=j+1

plt.xlabel("year")
plt.ylabel("HB < 7mg")
plt.legend()
plt.show()
```
